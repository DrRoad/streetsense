## Analyzing MTurk Coding



```python
import pandas as pd
```

```python
# Read in the files from geo sampling that are fed to Mturk
jakarta = pd.read_csv("../data/google_street_view_metadata/jakarta.csv")
lagos   = pd.read_csv("../data/google_street_view_metadata/lagos.csv")
bangkok = pd.read_csv("../data/google_street_view_metadata/bangkok.csv")
wayne   = pd.read_csv("../data/google_street_view_metadata/wayne2.csv")

# Read in the Mturk data
jakarta_mturk = pd.read_csv("../data/mturk/jakarta2_mturk_14_06_18.csv")  
lagos_mturk   = pd.read_csv("../data/mturk/lagos_mturk_2018_06_08.csv")
bangkok_mturk = pd.read_csv("../data/mturk/bangkok_mturk_2018_06_02.csv")
wayne_mturk   = pd.read_csv("../data/mturk/wayne_mturk_2018_06_11_18.csv")
```

```python
jakarta_mturk.head()
```

```python
# Remove HITs that were rejected from the MTurk files
print("Number of rows in the original Jakarta Mturk data: " + str(jakarta_mturk.shape[0]))
jakarta_mturk = jakarta_mturk.loc[jakarta_mturk['AssignmentStatus'] != "Rejected"]
print("After removing rejected HITs: " + str(jakarta_mturk.shape[0]))

print("Number of rows in the original Lagos Mturk data: " + str(lagos_mturk.shape[0]))
lagos_mturk = lagos_mturk.loc[lagos_mturk['AssignmentStatus'] != "Rejected"]
print("After removing rejected HITs: " + str(lagos_mturk.shape[0]))

print("Number of rows in the original Bangkok Mturk data: " + str(bangkok_mturk.shape[0]))
bangkok_mturk = bangkok_mturk.loc[bangkok_mturk['AssignmentStatus'] != "Rejected"]
print("After removing rejected HITs: " + str(bangkok_mturk.shape[0]))

print("Number of rows in the original Wayne Mturk data: " + str(wayne_mturk.shape[0]))
wayne_mturk = wayne_mturk.loc[wayne_mturk['AssignmentStatus'] != "Rejected"]
print("After removing rejected HITs: " + str(wayne_mturk.shape[0]))
```

```python
# Left join original data with Mturk
jakarta_all = jakarta.merge(jakarta_mturk, how = 'left', left_on = 'url_img0', right_on = 'Input.url_img0')
lagos_all   = lagos.merge(lagos_mturk, how = 'left', left_on = 'url_img0', right_on = 'Input.url_img0')
bangkok_all = bangkok.merge(bangkok_mturk, how = 'left', left_on = 'url_img0', right_on = 'Input.url_img0')
wayne_all   = wayne.merge(wayne_mturk, how = 'left', left_on = 'url_img0', right_on = 'Input.url_img0')
```

```python
# Number of rows in the final data (confirming)
print(jakarta_all.shape[0])
print(lagos_all.shape[0])
print(bangkok_all.shape[0])
print(wayne_all.shape[0])
```

```python
# Let's subset the data to relevant cols
small_subset = ('osm_type',
                'date',
                'Answer.Do you see any of the following?',
                'Answer.Does the road have any...?',
                'Answer.If there are any road markings are they clear?')
jakarta_small = jakarta_all.loc[:, small_subset]
lagos_small   = lagos_all.loc[:, small_subset]
bangkok_small = bangkok_all.loc[:, small_subset]
wayne_small   = wayne_all.loc[:, small_subset]

# Add a column enumerating city
jakarta_small['city'] = 'jakarta'
lagos_small['city']   = 'lagos'
bangkok_small['city'] = 'bangkok'
wayne_small['city']   = 'wayne'
```

```python
# Recoding for convenience and clarity

# Let's tally proportion of roads with potholes and cracks
jakarta_small['cracks'] = jakarta_small['Answer.Does the road have any...?'].str.contains('Cracks', na = False)
lagos_small['cracks']   = lagos_small['Answer.Does the road have any...?'].str.contains('Cracks', na = False)
bangkok_small['cracks'] = bangkok_small['Answer.Does the road have any...?'].str.contains('Cracks', na = False)
wayne_small['cracks']   = wayne_small['Answer.Does the road have any...?'].str.contains('Cracks', na = False)

jakarta_small['potholes'] = jakarta_small['Answer.Does the road have any...?'].str.contains('Potholes', na = False)
lagos_small['potholes']   = lagos_small['Answer.Does the road have any...?'].str.contains('Potholes', na = False)
bangkok_small['potholes'] = bangkok_small['Answer.Does the road have any...?'].str.contains('Potholes', na = False)
wayne_small['potholes']   = wayne_small['Answer.Does the road have any...?'].str.contains('Potholes', na = False)

# Road Markings
jakarta_small['clear_road_markings'] = jakarta_small['Answer.If there are any road markings are they clear?'] == 'Yes'
lagos_small['clear_road_markings']   = lagos_small['Answer.If there are any road markings are they clear?'] == 'Yes'
bangkok_small['clear_road_markings'] = bangkok_small['Answer.If there are any road markings are they clear?'] == 'Yes'
wayne_small['clear_road_markings']   = wayne_small['Answer.If there are any road markings are they clear?'] == 'Yes'

jakarta_small['roads_w_markings']    = jakarta_small['Answer.If there are any road markings are they clear?'].str.contains('Yes|No', na = False)
lagos_small['roads_w_markings']      = lagos_small['Answer.If there are any road markings are they clear?'].str.contains('Yes|No', na = False)
bangkok_small['roads_w_markings']    = bangkok_small['Answer.If there are any road markings are they clear?'].str.contains('Yes|No', na = False)
wayne_small['roads_w_markings']      = wayne_small['Answer.If there are any road markings are they clear?'].str.contains('Yes|No', na = False)

# Questions about litter on the street and whether the sidewalk was paved was not asked for Jakarta
jakarta_small['litter'] = jakarta_small['Answer.Do you see any of the following?'].str.contains('Litter', na = False)
lagos_small['litter']   = lagos_small['Answer.Do you see any of the following?'].str.contains('Litter', na = False)
bangkok_small['litter'] = bangkok_small['Answer.Do you see any of the following?'].str.contains('Litter', na = False)
wayne_small['litter']   = wayne_small['Answer.Do you see any of the following?'].str.contains('Litter', na = False)

jakarta_small['paved_sidewalk'] = jakarta_small['Answer.Do you see any of the following?'].str.contains('sidewalk', na = False)
lagos_small['paved_sidewalk']   = lagos_small['Answer.Do you see any of the following?'].str.contains('sidewalk', na = False)
bangkok_small['paved_sidewalk'] = bangkok_small['Answer.Do you see any of the following?'].str.contains('sidewalk', na = False)
wayne_small['paved_sidewalk']   = wayne_small['Answer.Do you see any of the following?'].str.contains('sidewalk', na = False)

# Let's subset on relevant cols.
fin_cols = ('city', 'osm_type', 'date', 'potholes', 'cracks', 'clear_road_markings', 'roads_w_markings', 'litter', 'paved_sidewalk')
jakarta_final = jakarta_small.loc[:, fin_cols]
lagos_final   = lagos_small.loc[:, fin_cols]
bangkok_final = bangkok_small.loc[:, fin_cols]
wayne_final   = wayne_small.loc[:, fin_cols]

# Concatenate
all_cities = pd.concat([jakarta_final, lagos_final, bangkok_final, wayne_final])
# Coerce data types
all_cities[['litter','paved_sidewalk']] = all_cities[['litter','paved_sidewalk']].apply(pd.to_numeric)
```

```python
from datetime import datetime
all_cities['date'] = pd.to_datetime(all_cities['date'])
all_cities.groupby(['city'])['date'].agg(['min', 'max'])
```

```python
# Let's do some group by means
all_cities.groupby(['city']).mean().round(2)
```

```python
# Proportion of road segments with markings that are clear
(all_cities.groupby(['city'])['clear_road_markings'].sum()/all_cities.groupby(['city'])['roads_w_markings'].sum()).round(2)
```

```python
all_cities.groupby(['city', 'osm_type']).mean().round(2)
```

```python
from sklearn.linear_model import LinearRegression

X_train = all_cities.loc[:, ('osm_type', 'city')]
y_potholes = all_cities[['potholes']]
y_cracks = all_cities[['cracks']]

X_train_fin = pd.get_dummies(X_train)
X_train_finn = X_train_fin.loc[:, ('osm_type_primary', 'osm_type_secondary', 'osm_type_trunk', 'city_bangkok', 'city_jakarta', 'city_lagos')]

print("Potholes:\n")
lm = LinearRegression()
lm.fit(X_train_finn, y_potholes)

for idx, col_name in enumerate(X_train_finn.columns):
    print("The coefficient for {} is {}".format(col_name, lm.coef_[0][idx].round(2)))

print("\nCracks:\n")
lm = LinearRegression()
lm.fit(X_train_finn, y_cracks)

for idx, col_name in enumerate(X_train_finn.columns):
    print("The coefficient for {} is {}".format(col_name, lm.coef_[0][idx].round(2)))
```
