### Join Gold Standard Answers with Turk Answers

```python
import pandas as pd
```

```python
# Turk
lagos_turk = pd.read_csv("../data/mturk_new/lagos_mturk_2019_03_19.csv")
bangkok_turk = pd.read_csv("../data/mturk_new/bangkok_mturk_2019_03_11.csv")
jakarta_turk = pd.read_csv("../data/mturk_new/jakarta_mturk_2019_02_22.csv")

turk_all = pd.concat([lagos_turk, bangkok_turk, jakarta_turk])

# KO
lagos_ko = pd.read_csv("../data/ko/ko_lagos.csv")
lagos_ko = lagos_ko.iloc[:,range(0,10)]
lagos_ko['url'] = lagos_ko['url'].astype(int)
lagos_ko.loc[lagos_ko['url'] < 10, 'url2'] =  'lagos/img0_000' + lagos_ko['url'].astype(str) + '.jpg'
lagos_ko.loc[(lagos_ko['url'] < 100) & (lagos_ko['url'] >= 10), 'url2'] =  'lagos/img0_00' + lagos_ko['url'].astype(str) + '.jpg'
lagos_ko.loc[lagos_ko['url'] >= 100, 'url2'] =  'lagos/img0_0' + lagos_ko['url'].astype(str) + '.jpg'

bangkok_ko = pd.read_csv("../data/ko/ko_bangkok.csv")
bangkok_ko = bangkok_ko.iloc[:,range(0,10)]
bangkok_ko.loc[bangkok_ko['url'] < 10, 'url2'] =  'bangkok/img0_000' + bangkok_ko['url'].astype(str) + '.jpg'
bangkok_ko.loc[(bangkok_ko['url'] < 100) & (bangkok_ko['url'] >= 10), 'url2'] =  'bangkok/img0_00' + bangkok_ko['url'].astype(str) + '.jpg'
bangkok_ko.loc[bangkok_ko['url'] >= 100, 'url2'] =  'bangkok/img0_0' + bangkok_ko['url'].astype(str) + '.jpg'

jakarta_ko = pd.read_csv("../data/ko/ko_jakarta.csv")
jakarta_ko = jakarta_ko.iloc[:,range(0,10)]
jakarta_ko.loc[jakarta_ko['url'] < 10, 'url2'] =  'jakarta/img0_000' + jakarta_ko['url'].astype(str) + '.jpg'
jakarta_ko.loc[(jakarta_ko['url'] < 100) & (jakarta_ko['url'] >= 10), 'url2'] =  'jakarta/img0_00' + jakarta_ko['url'].astype(str) + '.jpg'
jakarta_ko.loc[jakarta_ko['url'] >= 100, 'url2'] =  'jakarta/img0_0' + jakarta_ko['url'].astype(str) + '.jpg'

ko_all = pd.concat([lagos_ko, bangkok_ko, jakarta_ko])
```

```python
print(list(ko_all))
print(list(turk_all))
```

```python
ko_turk = ko_all.merge(turk_all, left_on='url2', right_on='Input.url_img0', how = 'left')
```

```python
ko_turk.columns = [c.lower().replace(' ', '_') for c in ko_turk.columns]
ko_turk.rename(columns={"answer.do_you_see_any_of_the_following?.a_paved_sidewalk": "turk_sidewalk",
                       "answer.do_you_see_any_of_the_following?.litter": "turk_litter",
                       "answer.radios.no_markings": "turk_no_markings",
                       "answer.radios.not_clear": "turk_not_clear"}, inplace = True)
```

```python
pd.crosstab(ko_turk.sidewalk, ko_turk.turk_sidewalk, margins=True)
```

```python
pd.crosstab(ko_turk.litter, ko_turk.turk_litter, margins=True)
```

```python
pd.crosstab(ko_turk.no_markings, ko_turk.turk_no_markings, margins=True)
```

```python
ko_turk['turk_cracks'] = ko_turk['answer.annotatedresult.boundingboxes'].apply(lambda c: sum(b['label']=='Crack' for b in eval(c)))
pd.crosstab(ko_turk.cracks, ko_turk.turk_cracks, margins=True)
```

```python
ko_turk['turk_pothole'] = ko_turk['answer.annotatedresult.boundingboxes'].apply(lambda c: sum(b['label']=='Pothole' for b in eval(c)))
pd.crosstab(ko_turk.pothole, ko_turk.turk_pothole, margins=True)
```
